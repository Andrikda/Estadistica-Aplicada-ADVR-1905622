```python
# Las de siempre
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# librerias de supervivencia
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test
from lifelines import CoxPHFitter
```

```python
# Datos
relacion = pd.read_csv("https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/km_relacion.csv")
```

```python
relacion.head()
```

La base de datos consiste en 42 observaciones. Es de un experimento, donde se dividio a las observaciones enh dos grupos. Es decir, cada uno de 21 observaciones.

**VARIABLES**:

**Tratamiento**: 1 le toco ser detallistas (dar un regalo en el transcurso de la semana) dentro en la, 0 no son detallistas.

**t**: tiempo de la relación en semanas.

**Estados**: refiere a 1 si se termino la relación y 0 si esta censurado.

**HIPÓTEIS**

La relación dura más si las personas son detallistas respectivamente con su pareja.

```python
#Estadisticas descriptivas
relacion.describe()
```

```python
relacion.groupby("tratamiento").agg(["mean", "std", "count"])
```

```python
detallistas = relacion.query("tratamiento == 1")
```

```python
no_detallistas= relacion.query("tratamiento == 0")
```

```python
detallistas
```

```python
no_detallistas
```

```python
km_d = KaplanMeierFitter().fit(detallistas["t"], detallistas["estado"])
km_nd = KaplanMeierFitter().fit(no_detallistas["t"], no_detallistas["estado"])
```

```python
km_d.plot_survival_function(label="detallistas")
km_nd.plot_survival_function(label="no detallistas")
plt.show()
```

Cuando de interceptan los intervalos se hace lo siguiente:

```python
lrt = logrank_test(detallistas["t"],detallistas["estado"],
                   no_detallistas["t"], no_detallistas["estado"])
lrt.print_summary()
```

### H0: funciones supervivencia iguales y  Ha: Funciones de supervivencias dsitintas

A un nivel de confianza del 95% se rechaza H0, por tanto las funciones de supervivencia de detallistas y no detallistas son distintas.

```python
cph = CoxPHFitter().fit(relacion,"t","estado")
cph.print_summary();
```

**Ser detallista disminuye el riesgo de que la relación termine, 1-exp(coef) = 1-.021 = 0.79**

**En promedio alguien que es detallista tiene 79% menos riesgo de que termine la relación respecto a alguien que no lo es detallista**
