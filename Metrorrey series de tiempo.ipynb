```python
import numpy as nop
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# especializadas
from statsmodels.tsa.api import seasonal_decompose
```

```python
# datos
metrorrey = pd.read_csv('metrorrey.csv')
```

```python
metrorrey.head()
```

```python
## vamos a definir como tiempo a la varible fecha
metrorrey['fecha'] = pd.to_datetime(metrorrey['fecha'], format='%Y/%m') #como sql es tipo fecha
```

```python
metrorrey.tail() #formato diario
```

```python
# la variable de fecha tenemos que convertirla en el index
metrorrey = metrorrey.set_index('fecha')
```

```python
# definir la frecuencia
metrorrey.head()
```

```python
#definir frecuenia "MS" en python es mensual
metrorrey.index.freq = 'MS'
```

```python
### con esto ya podemos hacer analisis de series
## confirmar
metrorrey.index
```

```python
metrorrey['Pasajeros'].plot();
```

```python
metrorrey['Ingresos'].plot();
```

```python
seried = seasonal_decompose(metrorrey['Pasajeros'], model='ad')
seried.plot();
```
