```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
x = [1,2,3,4,5]
y = [1,8,27,64,125]
```

```python
plt.plot(x,y) #Grafica de linea
```

```python
plt.bar(x,y) #Grafica de barras
```

```python
plt.scatter(x,y) #Grafica de puntos
```

```python
plt.barh(x,y) #Grafica de barras horizontales
```

```python
plt.plot(x,y,color="red")
```

```python
## Creamos dataframe

mun_nl = pd.DataFrame({"municipio": ["Monterrey","Guadalupe","Apodaca","San Nicolas","Santiago"], "nacimientos": [500,630,768,400,250]})
```

```python
mun_nl
```

```python
plt.bar(mun_nl["municipio"],mun_nl["nacimientos"])
plt.show()
```

```python
mun_nl = mun_nl.sort_values("nacimientos",ascending = False)
```

```python
plt.bar(mun_nl["municipio"],mun_nl["nacimientos"])
plt.show()
```

```python
#Grafica simple
plt.bar(mun_nl["municipio"],mun_nl["nacimientos"], color=["blue", "green", "red", "yellow", "gray"])
#Agregarrle titulo
plt.title("Nacimientos por municipio")
#Agregar nombre al eje x
plt.xlabel("Municipios de NL")
#Agregar nombre al eje y
plt.ylabel("Cantidad de nacimientos")
# Cambiar los saltos escala y
plt.yticks([0,200,400,600,800])
# Mostrar Grafica
plt.show()
```

```python
#fig es nombre del lienzo
fig, (panel1, panel2)=plt.subplots(2,1) #subgraficas, su nombre
fig.tight_layout()

#Panel1 (subgrafica 1)
x1=x #eje x de grafico 1
y1=y #eje y de grafico 1
panel1.plot(x,y, label="nombre de lineas")
panel1.title.set_text("Grafica1")

#Panel 2 (subgrafica 1)
x2=mun_nl["municipio"] #eje x de grafico 2
y2=mun_nl["nacimientos"] #eje y de grafico 2
panel2.bar(x2,y2)
panel2.title.set_text("Grafica2")

plt.show()
```

```python
mun_nl = mun_nl.sort_values("nacimientos",ascending = True)
#Grafica simple
plt.barh(mun_nl["municipio"],mun_nl["nacimientos"], color=["blue", "green", "red", "yellow", "gray"])
#Agregarrle titulo
plt.title("Nacimientos por municipio")
#Agregar nombre al eje x
plt.ylabel("Municipios de NL")
#Agregar nombre al eje y
plt.xlabel("Cantidad de nacimientos")
# Cambiar los saltos escala y
plt.xticks([0,200,400,600,800])
# Mostrar Grafica
plt.show()
```

```python
plt.scatter(x,y)
#Agregar nombre al grafico
plt.title("Grafica de puntos")
#Agregar nombre al eje x
plt.xlabel("Valores de x")
#Agregar nombre al eje y
plt.ylabel("Valores de y")
# Mostrar Grafica
plt.show()
```
