# Pronostico de la inlfaci칩n de EEUU

```python
#las de siempre
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Las de series
# Las de series de tiempo
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import adfuller

## Modelo ARIMA
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Los datos
usa = pd.read_excel("C:/Users/Lenovo/Downloads/ANDRIK-ESTADIS.APLICADA/USAinflation.xls")
```

```python
#Vamos a declarar el como variable de tiempo
usa["mes"]=pd.to_datetime(usa["mes"])
```

```python
#Vamos a hacer a la variable de tiempo index
usa = usa.set_index("mes")
```

```python
# Asignar la frecuencia
usa = usa.asfreq("MS")
```

```python
# Corroborar
usa.index
```

```python
# Comenzar con el an치lisis de la serie
usa["inflacion_usa"].plot(figsize=(12,6))
plt.title("Inflaci칩n Mensual de EEUU 2004-2023")
plt.show()
## fuente BLS USA
```

Esta serie no muestra tendencia (Inflaci칩n estable),parece caminata aleatoria (No estacionaria).

1. Probar si la serie es estacionaria

```python
#Prueba de dickey-fuller
ADF_test1 = adfuller(usa["inflacion_usa"])
# Ho: Serie No estacionaria

#Significancia 5%
alpha = 0.05

if ADF_test1[1] < alpha:
    print("Rechazamos Ho, La serie es estacionaria")
else:
    print("No rechazamos Ho, La serie es NO estacionaria")
```

Como la serie es no estacionario, aplicamos primeras diferencias

```python
usa["inflacion_diff"] = usa["inflacion_usa"].diff()
```

```python
# Graficar la inflaci칩n diferenciada
usa["inflacion_diff"].plot(figsize=(12,6))
plt.title("Primeras diferencias de la inflacion de EUU")
plt.show()
```

```python
# hacemos la prueba de Raiz Unitaria
#Prueba de dickey-fuller
ADF_test2 = adfuller(usa["inflacion_diff"].dropna())
# Ho: Serie No estacionaria

#Significancia 5%
alpha = 0.05

if ADF_test2[1] < alpha:
    print("Rechazamos Ho, La serie es estacionaria")
else:
    print("No rechazamos Ho, La serie es NO estacionaria")
```

Ya sabemos que es estacionaria, ahora debemos determinar que tipo de proceso es

```python
ig, ax = plt.subplots(1, 2, figsize=(12,4))

plot_acf(usa['inflacion_diff'].dropna(), zero=False, lags=40, ax=ax[0],title="Funci칩n de Autocorrelaci칩n")
ax[0].set_ylim(-1,1)
plot_pacf(usa['inflacion_diff'].dropna(), zero=False, lags=40, ax=ax[1], title="Funci칩n de Autocorrelaci칩n Parcial")
ax[1].set_ylim(-1,1)
plt.show()
```

Con base en los correlogramas podemos establecer que la inflaci칩n tiene un efecto estacional

```python
#Vamos a estimar varios modelos y que se guarde el AIC BIC
ic = {}
for ar in range(10):
    for ma in range(10):
        print(f"AR:{ar}, MA:{ma}")
        modelo = ARIMA(usa["inflacion_usa"].dropna(), order=(ar,1,ma))
        resultado = modelo.fit()
        ic[(ar,ma)] = [resultado.aic, resultado.bic]
        
ic = pd.DataFrame(ic, index=["AIC", "BIC"]).T
ic.index = ic.index.set_names(["AR","MA"])
```

```python
ic.sort_values('AIC', ascending=True).head(3)
```

De acuerdo con el AIC la inflaci칩n_usa es ARIMA(8,1,9)

```python
# Estimar el verdadero modelo
modelo_final = ARIMA(usa['inflacion_usa'], order=(8,1,9))
resultadof = modelo_final.fit()
print(resultadof.summary())
```

```python
# Pron칩stico
resultadof.predict('2023-04-01', '2023-07-01')
```

Para el mes de abril nuestro ARIMA pronostica una inflaci칩n de 8.97%

```python
# Librer칤a para la gr치fica del pron칩stico
from statsmodels.graphics.tsaplots import plot_predict

## Gr치fica
plot_predict(resultadof, start='2023-04-01', end='2023-07-01');
usa['inflacion_usa'].plot()
plt.xlim('2020-01-01', '2023-07-01');
```
